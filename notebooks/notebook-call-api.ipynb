```python
from dotenv import load_dotenv
import os

# Load variables from .env
load_dotenv()
token = os.getenv("AQICN_TOKEN")
print(token)
```

```python
import requests
import json


# API URL（utilise mon propre token）
url = "https://api.waqi.info/feed/france/?token=" + token

response = requests.get(url)


if response.status_code == 200:
    data = response.json()
    print("data : ", data)  
    if data['status'] == 'ok':
        print("Air quality data：")
        print(json.dumps(data['data'], indent=2, ensure_ascii=False))  
    else:
        print("faux API ：", data['status'])
else:
    print("❌ ：", response.status_code)
```

```python

url = "https://api.waqi.info/feed/here/?token=e42048adfee8ab48ec81057ddd672e13fc5ef36f"

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    if data['status'] == 'ok':
        aqi_data = data['data']

        # choisir les variables à afficher
        record = {
            "city": aqi_data['city']['name'],
            "aqi": aqi_data['aqi'],
            "time": aqi_data['time']['s'],
        }

        # KPI - AQI
        iaqi = aqi_data.get('iaqi', {})
        for key, value in iaqi.items():
            record[key] = value.get('v')

        # put DataFrame
        df = pd.DataFrame([record])
        print(df)

    else:
        print("⚠️ AQICN API status error:", data['status'])
else:
    print("❌ HTTP error:", response.status_code)
```
